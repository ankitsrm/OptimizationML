# **Eigen value decomposition : **

```python
def plotVecotrs(vecs,cols, alpha=1):
  """
  plot set of vectors 
    Parameters
    ---------
    VECS : array-like :: coordinate of the vectors to plot. Each vectors is in an array. for instance :[[1,2],[2,2],[3,2]] can be used to plot 2 vectors 
    cols : array like colors of the vectors. for instance: ['red','blue'] will display the first vectors in red and second vector in blue.
    alpha: float 
      Opactity of vectors

    RETURNS :
      fig :instance of matplotlib.figur.Figure
        the Figure of the vectors.
  """
  plt.figure()
  plt.axvline(x=0,color='#A9A9A9', zorder=0)
  plt.axhline(y=0, color='#A9A9A9', zorder=0)

  for i in range(len(vecs)):
    x= np.concatenate([[0,0], vecs[i]])
    plt.quiver([x[0]],
                [x[1]], 
                 [x[2]], 
                  [x[3]],
                   angles='xy', scale_units='xy',scale=1, color=cols[i],alpha=alpha)
    
```

We are defining a matrix and a vector

```python
A = np.array([[-1,3],[2,-2]])
V = np.array([[2],[2]])
```
